```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.random.randn(10000)
y = np.random.randn(10000)
```

```python
plt.scatter(x,y)
plt.xlim(-5,5)
plt.ylim(-5,5)
plt.title('Gaussian data')
```

use fig

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(x,y)
ax.set_xlim(0,4)
ax.set_ylim(0,4)
plt.show()
```
