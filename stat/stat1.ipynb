```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, axes = plt.subplots(2,2)

fig.set_figwidth(16)

sample = np.random.randn(100000)

axes[0][0].plot(sorted(sample))
axes[1][0].hist(sample, 100)

sample = np.random.random(100000) - 0.5

axes[0][1].plot(sorted(sample))
axes[1][1].hist(sample, 100)

axes[0][1].set_ylim([-1,1])
axes[1][1].set_ylim([0,4000])

plt.show()
```
