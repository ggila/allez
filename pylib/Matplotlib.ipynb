```python
%matplotlib inline
import numpy as np
```

```python
x = np.linspace(0,1,11)
```

```python
plt.plot(x, x, alpha=0.7)
plt.plot(x, x**2, 'k+-')
ax = plt.gca()
ax.set_xlim([0,1])
line = ax.lines[0]
line.set_marker('o')
ax.annotate('middle', xy=(50, 50), xytext=(20,70))
```
