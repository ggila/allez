```python
def sincos(a,b):
    plt.plot(x, np.cos(x), 'k', label='$\cos(x)$')
    plt.plot(x, a*np.sin(x+b), 'r', label='$a*\sin(x+b)$')
    plt.xlim(-2*np.pi, 2*np.pi)
    plt.ylim(-2,2)

x = np.linspace(-2*np.pi, 2*np.pi)
```

```python
sincos(1.4,2)
```
