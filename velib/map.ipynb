## basemap

install: http://matplotlib.org/basemap/users/installing.html    
doc: http://matplotlib.org/basemap/api/basemap_api.html#module-mpl_toolkits.basemap

```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
```

This is what latitude and longitude looks like:

<img src='latitude-longitude.jpg'>

```python
m = Basemap(projection='merc',llcrnrlat=-80,urcrnrlat=80,\
            llcrnrlon=-180,urcrnrlon=180)

plt.figure(figsize=(15,15))

m.drawcoastlines()
m.fillcontinents(color='white',lake_color='brown')


m.drawparallels([48.8567], 'r')
m.drawmeridians([2.3508], 'r')
m.drawmapboundary(fill_color='black')

plt.title("Mercator Projection")
```

Paris: 48.8567° N, 2.3508° E

```python
m = Basemap(projection='merc',llcrnrlat=38,urcrnrlat=60,\
            llcrnrlon=-17,urcrnrlon=17)

plt.figure(figsize=(15,15))
m.drawcoastlines()
m.drawrivers(color='blue')
m.drawcountries(linewidth=2)

m.drawparallels([48.8567], 'r')
m.drawmeridians([2.3508], 'r')



plt.title("western europe")
```

```python
m = Basemap(projection='merc',llcrnrlat=41,urcrnrlat=53,\
            llcrnrlon=-6,urcrnrlon=10)
m.bluemarble()
m.drawcountries(linewidth=2)
#m.drawcounties()
plt.title("france")
```

## Velib

We can download station data from:
https://developer.jcdecaux.com/#/home

```python
plt.figure(figsize=(15,15))

m = Basemap(projection='merc',llcrnrlat=48.80,urcrnrlat=48.93,\
            llcrnrlon=2.21,urcrnrlon=2.49)

paris = pd.read_csv('Paris.csv')


for lat, lon in zip(paris.Latitude, paris.Longitude):
    x, y = m(lon, lat)
    m.plot(x, y, 'bo', alpha=0.4)
    
```
